```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# The csv file for Spotify Most Streamed Songs will be referred to as 'data' for coding
data=pd.read_csv("Spotify Most Streamed Songs.csv", encoding="ISO-8859-1")
data.head()
```

## Dataset features
#### Basic Track Information:
- track_name: Name of the song.
- artist(s)_name: Name of the artist(s) performing the song.
- artist_count: Number of artists contributing to the song.
- released_year, released_month, released_day: Release date details.
#### Streaming Metrics:
- in_spotify_playlists: Number of Spotify playlists the song is featured in.
- in_spotify_charts: Rank of the song on Spotify charts.
- streams: Total number of streams on Spotify.
- in_apple_playlists, in_apple_charts: Presence in Apple Music playlists and charts.
- in_deezer_playlists, in_deezer_charts: Presence in Deezer playlists and charts.
- in_shazam_charts: Rank on Shazam charts.
#### Musical Attributes:
- bpm: Beats per minute, representing the tempo of the song.
- key: Key of the song.
- mode: Indicates whether the song is in a major or minor mode.
- danceability_%: Suitability of the song for dancing.
- valence_%: Positivity of the song’s musical content.
- energy_%: Perceived energy level of the song.
- acousticness_%: Acoustic sound presence in the song.
- instrumentalness_%: Proportion of instrumental content in the track.
- liveness_%: Presence of live performance elements.
- speechiness_%: Amount of spoken words in the song. spoken words in the song.

# 4. Data Set Graphical Exploration

```python
# Scatter plot of songs by BPM
sns.scatterplot(data['bpm'])
plt.xlabel("Number of Songs")
plt.title("Songs by BPM")
```

```python
# Danceability box plot
sns.boxplot(y=data["danceability_%"],x=data["bpm"])
plt.title("Danceability of songs by BPM")
# This isn't very useful as there is too wide a spread of BPM across all the songs
```

```python
# Histogram of Spotify Chart
sns.histplot(data["in_spotify_charts"],kde=True)
plt.title("Distribution of Spotify Chart Positions")
```

```python
# Counting the number of songs for each month
monthly_counts = data['released_month'].value_counts().sort_index()

#Bar chart
plt.figure(figsize=(10, 10))
monthly_counts.plot(kind='bar')

plt.title('Number of Songs Released Each Month')
plt.xlabel('Month')
plt.ylabel('Number of Songs')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)
plt.show()
```

```python
#Creating a line plot to show the 
yearly_counts = data['released_year'].value_counts().sort_index()

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Number of Songs Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')

# Show the plot
plt.show()
```
